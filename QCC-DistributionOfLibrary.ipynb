```python
#% matplotlib notebook 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/adrishg/LR4-repo/main/A3_lib-_adriana-T7-Term.tsv", sep='\t')
data
```

```python
def makeQCCDistributions(Position):
    for i in data['Position']:
        if i==Position:
            realPos = i-1 #Patch
            sizes= [data['A Value'][realPos], data['C Value'][realPos], data['G Value'][realPos],data['T Value'][realPos]]
            #print(sizes)
            tags = 'A Value', 'C Value', 'G Value', 'T Value'
            

    fig, ax = plt.subplots()
    ax.pie(sizes, labels=tags)


#makeQCCDistributions(3)
```

```python
#Distribution of G152: Position 131,132,133
makeQCCDistributions(131)
makeQCCDistributions(132)
makeQCCDistributions(133)
```

```python
#Distribution of A3: 580,579,578
makeQCCDistributions(580)
makeQCCDistributions(579)
makeQCCDistributions(578)
```
